Informasi Column
1. hotel -> Jenis hotel (misalnya City Hotel atau Resort Hotel).

2. is_canceled ->Status pembatalan reservasi (1 = dibatalkan, 0 = tidak dibatalkan)

3. lead_time -> Jumlah hari antara tanggal pemesanan dan tanggal kedatangan.

4. arrival_date_year -> Tahun kedatangan tamu.

5. arrival_date_month -> Bulan kedatangan tamu (dalam bentuk teks, misalnya January).

6. arrival_date_week_number -> Nomor minggu dalam tahun saat tamu tiba.

7. arrival_date_day_of_month -> Tanggal (hari ke-) kedatangan dalam satu bulan.

8. stays_in_weekend_nights -> Jumlah malam menginap pada akhir pekan (Sabtu–Minggu).

9. stays_in_week_nights -> Jumlah malam menginap pada hari kerja (Senin–Jumat).

10. adults -> Jumlah orang dewasa dalam reservasi.

11. children -> Jumlah anak-anak dalam reservasi (ada nilai kosong/null).

12. babies -> Jumlah bayi dalam reservasi.

13. meal -> Jenis paket makanan yang dipesan (misalnya BB, HB, FB, SC).

14. country -> Kode negara asal tamu (ISO country code).

15. market_segment -> Segmen pasar pemesanan (misalnya Online TA, Direct, Corporate).

16. distribution_channel -> Saluran distribusi pemesanan (misalnya TA/TO, Direct).

17. is_repeated_guest -> Menunjukkan apakah tamu adalah tamu lama (1 = tamu berulang, 0 = tamu baru)

18. previous_cancellations -> Jumlah pemesanan sebelumnya yang dibatalkan oleh tamu.

19. previous_bookings_not_canceled -> Jumlah pemesanan sebelumnya yang tidak dibatalkan.

20. reserved_room_type -> Tipe kamar yang dipesan saat reservasi.

21. assigned_room_type -> Tipe kamar yang benar-benar diberikan kepada tamu.

22. booking_changes -> Jumlah perubahan yang dilakukan pada reservasi.

23. deposit_type -> Jenis deposit (No Deposit, Non Refund, Refundable).

24. agent -> ID agen perjalanan yang melakukan pemesanan (banyak nilai kosong).

25. company -> ID perusahaan jika pemesanan dilakukan atas nama perusahaan.

26. days_in_waiting_list -> Jumlah hari reservasi berada dalam daftar tunggu.

27. customer_type -> Jenis pelanggan berdasarkan pola menginap, seperti Transient atau Group.

28. adr (Average Daily Rate) -> Rata-rata harga kamar per malam yang dibayarkan tamu.

29. required_car_parking_spaces -> Jumlah tempat parkir mobil yang diminta.

30. total_of_special_requests -> Jumlah permintaan khusus (misalnya ranjang bayi, kamar tertentu).

31. reservation_status -> Status akhir reservasi (Canceled, Check-Out, No-Show).

32. reservation_status_date -> Tanggal terakhir status reservasi diperbarui.

33. bookingID -> ID unik untuk setiap pemesanan.

```python
df = df.dropna()
```

```python
# Menampilkan Jumlah Data
df.shape
```

```python
# a. Mendeteksi Outlier (Menggunakan IQR)
Q1 = df[numerik].quantile(0.25)
Q3 = df[numerik].quantile(0.75)
IQR = Q3 - Q1

# Filter Baris Outlier
outlier_iqr = df[((df[numerik] < (Q1 - 1.5 * IQR)) | (df[numerik] > (Q3 + 1.5 * IQR))).any(axis = 1)]
print("Data Outlier :", outlier_iqr.shape[0], "Baris")
```

```python
import matplotlib.pyplot as plt

# Ambil hanya kolom numerik
numerik = df.select_dtypes(include=["int64", "float64"]).columns

# Buat boxplot
plt.figure(figsize=(14, 6))
df[numerik].boxplot(rot=90)

plt.title("Boxplot Deteksi Outlier pada Kolom Numerik")
plt.ylabel("Nilai")
plt.show()
```

```python
# Feature Engineering
```

```python
# Kategori Lead Time (Tanggal Pemesanan dan Kedatangan)
df["lead_time_category"] = pd.cut(
    df["lead_time"],
    bins=[-1, 30, 90, 180, df["lead_time"].max()],
    labels=["Pendek", "Menengah", "Panjang", "Sangat Panjang"]
)
```

```python
# Total Tamu 
df["total_guests"] = (df["adults"] + df["children"] + df["babies"]).astype(int)
```

```python
# Total Lama Menginap
df["total_stay"] = df["stays_in_weekend_nights"] + df["stays_in_week_nights"]
```

```python
# Kategori Harga (ADR)
df["adr_category"] = pd.qcut(
    df["adr"],
    q=4,
    labels=["Murah", "Sedang", "Mahal", "Premium"]
)
```

```python
# Indikator Perubahan Reservasi
df["has_booking_changes"] = df["booking_changes"].apply(lambda x: 1 if x > 0 else 0)
```

```python
# Visualisasi (Target is_canceled)
# Tingkat Pembatalan Reservasi (Overview)

import matplotlib.pyplot as plt

df["is_canceled"].value_counts().plot(kind="bar")
plt.title("Distribusi Status Pembatalan Reservasi")
plt.xlabel("Status (0 = Tidak Batal, 1 = Batal)")
plt.ylabel("Jumlah Reservasi")
plt.show()
```

```python
# Pembatalan Berdasarkan Lead Time(Jumlah Hari Pemesanan dengan Kedatangan)
plt.figure()
df.boxplot(column="lead_time", by="is_canceled")
plt.title("Lead Time vs Status Pembatalan")
plt.suptitle("")
plt.xlabel("Status Pembatalan")
plt.ylabel("Lead Time (hari)")
plt.show()
```

```python
# Pembatalan Berdasarkan Harga Kamar (ADR)
plt.figure()
df.boxplot(column="adr", by="is_canceled")
plt.title("ADR vs Status Pembatalan")
plt.suptitle("")
plt.xlabel("Status Pembatalan")
plt.ylabel("Average Daily Rate")
plt.show()
```

```python
# Pembatalan Berdasarkan Total Lama Menginap
df["total_stay"] = df["stays_in_weekend_nights"] + df["stays_in_week_nights"]

plt.figure()
df.boxplot(column="total_stay", by="is_canceled")
plt.title("Total Lama Menginap vs Pembatalan")
plt.suptitle("")
plt.xlabel("Status Pembatalan")
plt.ylabel("Total Malam Menginap")
plt.show()
```

```python
# Pembatalan Berdasarkan Tipe Pelanggan
df.groupby("customer_type")["is_canceled"].mean().plot(kind="bar")
plt.title("Rata-rata Pembatalan Berdasarkan Tipe Pelanggan")
plt.xlabel("Customer Type")
plt.ylabel("Tingkat Pembatalan")
plt.show()
```

```python
# Pembatalan Berdasarkan Segmen Pasar
df.groupby("market_segment")["is_canceled"].mean().plot(kind="bar")
plt.title("Tingkat Pembatalan Berdasarkan Market Segment")
plt.xlabel("Market Segment")
plt.ylabel("Tingkat Pembatalan")
plt.show()
```
